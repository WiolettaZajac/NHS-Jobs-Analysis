```python
# salary
df['salary'].head(10)
```

```python
# Remove the encoding issue
df.loc[:, 'salary'] = df['salary'].str.replace('Â£', '£', regex=False)
```

```python
df['salary'].head(10)
```

```python
# Replace non-numeric values with missing values
df.loc[:, 'salary_clean'] = df['salary'].replace(
    ['TBC', 'Competitive', 'Medical Ad Hoc'],
    np.nan
)
```

```python
df[['salary', 'salary_clean']].head(10)
```

```python
#Extract the first salary value
df.loc[:, 'salary_clean'] = df['salary_clean'].str.extract(
    r'([\d,]+)', expand=False
)
```

```python
df[['salary', 'salary_clean']].head(10)
```

```python
# remove £ sign
df.loc[:, 'salary_clean'] = (
    df['salary_clean']
    .str.replace('£', '', regex=False)
    .str.replace(',', '', regex=False)
)
```

```python
# convert to numbers
df.loc[:, 'salary_clean'] = pd.to_numeric(df['salary_clean'], errors='coerce')
```

```python
# histogram
plt.figure(figsize=(8,5))

plt.hist(df['salary_clean'].dropna(), bins=20)

plt.title('Distribution of Salaries')
plt.xlabel('Salary (£)')
plt.ylabel('Number of Jobs')

plt.show()
```

*The original salary column contained text values such as salary ranges, 'Competitive', and 'TBC'. I cleaned the data by removing non-numeric entries and extracting the first salary value from each record. I converted the cleaned salaries into numeric values for statistical analysis.*

```python
salary = df[df['salary_clean'] <= 150000]

plt.figure(figsize=(8,5))

counts, bins, patches = plt.hist(
    salary['salary_clean'],
    bins=20,
    color="steelblue",
    edgecolor="black"
)

# Add labels above each bar
for count, patch in zip(counts, patches):
    if count > 0:
        plt.text(
            patch.get_x() + patch.get_width() / 2,
            count,
            f'{int(count)}',
            ha='center',
            va='bottom',
            fontsize=9
        )

plt.title('Distribution of Salaries')
plt.xlabel('Salary (£)')
plt.ylabel('Number of Jobs')

plt.show()
```

Salary Histogram

The histogram shows that most NHS jobs have salaries in the lower and middle salary ranges. There are fewer jobs with very high salaries.

```python
# top 10 specialities
df['specialty'].value_counts().head(10)
```

```python
# horizontal chart
top_specialties = df['specialty'].value_counts().head(10)

plt.figure(figsize=(10,6))

bars = plt.barh(
    top_specialties.index,
    top_specialties.values,
    color='teal',
    edgecolor='black'
)

# Add labels
for bar in bars:
    width = bar.get_width()
    plt.text(
        width + 5,
        bar.get_y() + bar.get_height()/2,
        f'{int(width)}',
        va='center'
    )

# Put the largest value at the top
plt.gca().invert_yaxis()

plt.title('Top 10 NHS Specialties')
plt.xlabel('Number of Vacancies')
plt.ylabel('Specialty')

plt.tight_layout()
plt.show()
```

Top 10 Specialties

This chart shows the specialties with the highest number of job vacancies. It shows which areas of the NHS need the most employees.

```python
# Horizontal bar chart
top_specialties = df['specialty'].value_counts().head(10)

colors = [
    'teal',
    'steelblue',
    'skyblue',
    'cornflowerblue',
    'mediumseagreen',
    'cadetblue',
    'turquoise',
    'dodgerblue',
    'lightseagreen',
    'royalblue'
]

plt.figure(figsize=(10,6))

bars = plt.barh(
    top_specialties.index,
    top_specialties.values,
    color=colors,
    edgecolor='black'
)

# Add labels
for bar in bars:
    width = bar.get_width()
    plt.text(
        width + 5,
        bar.get_y() + bar.get_height()/2,
        f'{int(width)}',
        va='center'
    )

# Put the largest value at the top
plt.gca().invert_yaxis()

plt.title('Top 10 NHS Specialties')
plt.xlabel('Number of Vacancies')
plt.ylabel('Specialty')

plt.tight_layout()
plt.show()
```

```python
# TOP 5 working patterns in a pie chart

top_patterns = df['working_pattern'].value_counts().head(5)

colors = ['steelblue', 'lightblue', 'skyblue', 'cornflowerblue', 'lightsteelblue']

plt.figure(figsize=(7,7))

plt.pie(
    top_patterns,
    labels=top_patterns.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    wedgeprops={'edgecolor': 'white'}
)

plt.title('Top 5 Working Patterns')

plt.show()
```
